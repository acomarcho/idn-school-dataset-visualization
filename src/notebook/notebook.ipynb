```python
import pandas as pd
```

# Prepare dataframe

```python
dataframe = pd.read_csv('../../out/complete_data.csv')
dataframe.head()
```

# Perform aggregation by province_name

```python
province_school_counts = dataframe.groupby('province_name')
province_school_counts = province_school_counts.size().to_frame(name='school_count').reset_index()
province_school_counts = province_school_counts.sort_values(by='school_count', ascending=False)

province_school_counts
```

```python
stage_school_counts = dataframe.groupby('stage')
stage_school_counts = stage_school_counts.size().to_frame(name='school_count').reset_index()
stage_school_counts = stage_school_counts.sort_values(by='school_count', ascending=False)

stage_school_counts
```

```python
grouped_by_province = dataframe.groupby('province_name')

province_school_ratio = grouped_by_province.size().to_frame(name='school_count')
province_school_ratio['all_population'] = grouped_by_province['total_population'].max() # Every entry have the same total_population
province_school_ratio['education_age_population'] = grouped_by_province['total_education_age_population'].max() # Every entry have the same total_education_age_population
province_school_ratio['ratio_per_all_capita'] = province_school_ratio['school_count'] / grouped_by_province['total_population'].sum()
province_school_ratio['ratio_per_education_age_capita'] = province_school_ratio['school_count'] / grouped_by_province['total_education_age_population'].sum()
province_school_ratio = province_school_ratio.reset_index()

province_school_ratio
```
