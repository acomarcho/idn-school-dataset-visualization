```python
import pandas as pd
```

# Prepare dataframe

```python
dataframe = pd.read_csv('../../out/complete_data.csv')
dataframe.head()
```

# Perform aggregation by province_name

```python
grouped_by_province = dataframe.groupby('province_name')
grouped_by_province = grouped_by_province.size().to_frame(name='school_count').reset_index()
grouped_by_province = grouped_by_province.sort_values(by='school_count', ascending=False)

grouped_by_province
```
